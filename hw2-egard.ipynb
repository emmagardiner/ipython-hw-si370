```python
df_groups = df_proximities.groupby('Group')
```

```python
df_groups['Protein'].mean() # Question 3(b)
```

```python
df_max_food_by_protein = df_groups.apply(lambda x: df_proximities.ix[np.argmax(x.Protein), ['Food','Protein']])
df_max_food_by_protein # Question 3(c) - by protein
```

```python
df_max_food_by_energy = df_groups.apply(lambda x: df_proximities.ix[np.argmax(x.Energy), ['Food','Energy']])
df_max_food_by_energy # Question 3(c) - by energy
```

```python
df_max_food_by_sugar = df_groups.apply(lambda x: df_proximities.ix[np.argmax(x.Sugars), ['Food','Sugars']])
df_max_food_by_sugar# Question 3(c) - by sugars
```

```python
df_max_food_by_water = df_groups.apply(lambda x: df_proximities.ix[np.argmax(x.Water), ['Food','Water']])
df_max_food_by_water # Question 3(c) - by water
```

```python
df_sweets = df_proximities[(df_proximities.Group == 'Sweets')][['Group','Energy']]
```

```python
df_fat_new = df_fat[['Group', 'Energy']]
```

```python
sweets = df_sweets.rename(columns ={'Group': 'Sweets', 'Energy':'Sweets-Energy'})
```

```python
fat = df_fat_new.rename(columns={'Group': 'Fast Food', 'Energy':'Fast-Food-Energy'})
```

```python
sweets_fats = fat.append(sweets)
```

```python
sweets_fats.ix[:, 'Fast-Food-Energy': 'Sweets-Energy'].boxplot(return_type="axes") # Question 3(g)
```

```python
from scipy.stats import ttest_ind
ttest_ind(df_fat.Energy, df_sweets.Energy, equal_var=False) # Question 3(h)
```

```python
df_proximities = df_proximities.drop('GroupID', 1)
```
